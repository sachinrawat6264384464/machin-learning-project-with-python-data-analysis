```python
import ast

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movie["genres"]=movie["genres"].apply(convert)
```

```python
movie.head(2)
```

```python
movie["keywords"]=movie["keywords"].apply(convert)
```

```python
def convert3(obj):
    L= []
    convert= 0
    for i in ast.literal_eval(obj):
        if convert != 3:
            L.append(i['name'])
            convert+=1
        else:
            break
        return L
```

```python
movie['cast']=movie['cast'].apply(convert3)
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movie["crew"]=movie["crew"].apply(fetch_director)
```

```python
movie["overview"]=movie["overview"].apply(lambda x:x.split())
```

```python
movie["genres"]=movie["genres"].apply(lambda x:[i.replace(" ","")for i in x])
movie["keywords"]=movie["keywords"].apply(lambda x:[i.replace(" ","")for i in x])

movie["crew"]=movie["crew"].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movie['tags'] = movie['overview'] + movie['genres'] + movie['keywords'] + movie['cast'] + movie['crew']
```

```python
movies = pd.read_csv("movies.csv",usecols=['movieId','title'])
movies.head(2)
```

```python
ratin = pd.read_csv("ratings.csv",usecols=['userId','movieId','rating'])
ratin.head(2)
```

```python
ratin.shape
movies.shape
```

```python
from fuzzywuzzy import process
movie_user=ratin.pivot(index='movieId',columns='userId',values='rating').fillna(0)
```

```python
from scipy.sparse import csr_matrix
```

```python
mat_movies = csr_matrix(movie_user.values)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(metric = 'cosine',algorithm = 'brute',n_neighbors = 10)
model.fit(mat_movies)
```

```python
def recommender(movie_name,data,n):
    idx = process.extractOne(movie_name,movies['title'])[2]
    print('movie selected :',movies['title'][idx],'Index :',idx)
    print('searching for recommondation.......')
    distance , indices = model.kneighbors(data[idx],n_neighbors=n)
    for i in indices:
       print(movies['title'][i].where(i!=idx))
```

```python
recommender('Trzy kolory',mat_movies,10)
```
